# 📊 Student Scores Data Analysis

This is a simple data analysis project I made to practice using Python and Pandas.
The dataset contains fictional student scores in 3 subjects.
I'll load the data, explore it a bit, calculate some basic stats, and make a chart.


## Step 1 — Import the libraries I need

```python
# pandas is for working with data (like Excel but in Python)
import pandas as pd

# matplotlib is for making charts
import matplotlib.pyplot as plt
```

## Step 2 — Create the dataset

Instead of loading a file, I'm creating the data directly in the code.
It has 10 students and their scores in Math, English, and Science.

```python
# I'm making the data as a dictionary first, then turning it into a DataFrame
data = {
    'Student': ['Ali', 'Sara', 'Lina', 'Omar', 'Nour',
                'Khalid', 'Reem', 'Yousef', 'Dina', 'Faris'],
    'Math':    [85, 92, 78, 65, 90, 55, 88, 72, 95, 60],
    'English': [76, 88, 82, 70, 91, 60, 74, 85, 89, 63],
    'Science': [90, 85, 75, 68, 87, 58, 80, 77, 93, 65]
}

# Create the DataFrame
df = pd.DataFrame(data)

print("Dataset created successfully!")
```

## Step 3 — Look at the data

First thing I always do is just look at what the data looks like.

```python
# .head() shows the first 5 rows
df.head()
```

```python
# .shape tells me how many rows and columns there are
print("Rows and columns:", df.shape)
```

```python
# .describe() gives a quick summary of the numbers (min, max, average, etc.)
df.describe()
```

## Step 4 — Calculate some basic stats

Now I want to find the average, highest, and lowest score for each subject.

```python
# Calculate average score per subject
avg_math    = df['Math'].mean()
avg_english = df['English'].mean()
avg_science = df['Science'].mean()

print("Average Scores:")
print(f"  Math:    {avg_math}")
print(f"  English: {avg_english}")
print(f"  Science: {avg_science}")
```

```python
# Find the highest score in Math and who got it
max_math_score   = df['Math'].max()
top_math_student = df.loc[df['Math'].idxmax(), 'Student']

print(f"Highest Math score: {max_math_score} (by {top_math_student})")
```

```python
# Find the lowest score in English and who got it
min_english_score    = df['English'].min()
lowest_english_student = df.loc[df['English'].idxmin(), 'Student']

print(f"Lowest English score: {min_english_score} (by {lowest_english_student})")
```

```python
# Add a new column for each student's average score across all 3 subjects
df['Average'] = df[['Math', 'English', 'Science']].mean(axis=1).round(1)

# Show the updated table
df
```

## Step 5 — Visualization

I'll make a bar chart showing the average score of each student.
This makes it easier to see who is doing better overall.

```python
# Bar chart — average score per student
plt.figure(figsize=(10, 5))

plt.bar(df['Student'], df['Average'], color='steelblue')

# Add labels and title
plt.title('Average Score per Student', fontsize=14)
plt.xlabel('Student')
plt.ylabel('Average Score')

# Add a horizontal line at the overall average so it's easier to compare
overall_avg = df['Average'].mean()
plt.axhline(y=overall_avg, color='red', linestyle='--', label=f'Class average: {overall_avg:.1f}')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Bonus: histogram of Math scores to see how they're distributed
plt.figure(figsize=(7, 4))

plt.hist(df['Math'], bins=5, color='coral', edgecolor='black')

plt.title('Distribution of Math Scores', fontsize=14)
plt.xlabel('Score')
plt.ylabel('Number of Students')

plt.tight_layout()
plt.show()
```
